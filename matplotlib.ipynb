1.Open source

2.Highly Customizable

3.Support 20+ chart types

```python
# First Graph using Matplotlib

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line Graph
plt.plot(
    x,
    y,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='blue',
    label='Sales'
)

# Title and Labels
plt.title("First Line Graph", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)


# Save the graph
plt.savefig("first_graph.png", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.jpg", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.pdf", bbox_inches="tight")

# Display the graph
plt.show()

# Close the figure
plt.close()
```
