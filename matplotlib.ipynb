```python
# First Graph using Matplotlib

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line Graph
plt.plot(
    x,
    y,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='blue',
    label='Sales'
)

# Title and Labels
plt.title("First Line Graph", fontsize=16)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Y Values", fontsize=12)


# Save the graph
plt.savefig("first_graph.png", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.jpg", dpi=300, bbox_inches="tight")
plt.savefig("first_graph.pdf", bbox_inches="tight")

# Display the graph
plt.show()

# Close the figure
plt.close()
```

**Histogram**

A histogram is a graphical representation used to display the frequency distribution of continuous numerical data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset

df = pd.read_csv("/content/house_prices.csv", on_bad_lines='skip', engine='python')

print("DataFrame Columns:", df.columns)

# =====================================================
# 1. Basic Histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(df["Price (in rupees)"]) # A histogram groups numerical data into intervals (bins) and displays the frequency.

plt.title("Basic Histogram")

plt.show()

# =====================================================
# 2. bins
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(                         # bins decides how many intervals (groups) the data is divided into.
    df["Price (in rupees)"],
    bins=20
)

plt.title("Histogram using Bins")

plt.show()

# =====================================================
# 3. density
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Instead of showing frequency, it shows the Probability Density.
    bins=20,
    density=True
)

plt.title("Density Histogram")

plt.show()

# =====================================================
# 4. cumulative histogram
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"],  # Shows the cumulative frequency.
    bins=20,
    cumulative=True
)

plt.title("Cumulative Histogram")

plt.show()

# =====================================================
# 5. transparency (alpha)
# =====================================================

plt.figure(figsize=(10,5))

plt.hist(
    df["Price (in rupees)"], # Transparency is controlled using the alpha parameter.
    bins=20,
    alpha=0.5
)

plt.title("Histogram using Transparency")

plt.show()
```

#Scatter Plot
Scatter plot is used to visualize the relationship between two numerical variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/Iris.csv")

# =====================================================
# 1. Basic scatter()
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"]
)

plt.title("Basic Scatter Plot")

plt.show()

# =====================================================
# 2. size (s). # The s parameter controls the size of the markers.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    s=100
)

plt.title("Scatter Plot with Size")

plt.show()

# =====================================================
# 3. color. # Changes the color of markers.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    color="red"
)

plt.title("Scatter Plot with Color")

plt.show()

# =====================================================
# 4. alpha. # Controls the transparency.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    alpha=0.5
)

plt.title("Scatter Plot with Transparency")

plt.show()

# =====================================================
# 5. marker. # Changes the shape of the points.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    marker="^"
)

plt.title("Scatter Plot with Marker")

plt.show()

# =====================================================
# 6. colormap (cmap) # A colormap assigns colors to points based on numeric values in c.
# =====================================================

plt.figure(figsize=(8,5))

plt.scatter(
    df["SepalLengthCm"],
    df["PetalLengthCm"],
    c=df["PetalWidthCm"],
    cmap="viridis",
    s=80
)

plt.colorbar(label="Petal Width")

plt.title("Scatter Plot with Colormap")

plt.show()
```

#Pie Chart
A Pie Chart is used to display the proportion or percentage of each category relative to the whole.

```python
# =============================================
# Module 6 - Pie Charts
# Dataset : titanic.csv
# Perform All Operations Individually
# =============================================

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/Iris.csv")

# Count of each species to create pie chart
survival = df["Species"].value_counts()

labels = survival.index.tolist()

# =====================================================
# 1. Basic pie()
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(survival.values)

plt.title("Basic Pie Chart")

plt.show()

# =====================================================
# 2. labels # Adds labels to each slice.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels
)

plt.title("Pie Chart with Labels")

plt.show()

# =====================================================
# 3. autopct # Displays percentage values on each slice.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    autopct="%1.1f%%"
)

plt.title("Pie Chart with Percentage")

plt.show()

# =====================================================
# 4. explode # explode is used to separate one or more slices from the pie chart.
# =====================================================

# Adjust explode to match the number of species in Iris dataset
explode = [0.1, 0, 0]

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    explode=explode,
    autopct="%1.1f%%"
)

plt.title("Pie Chart using Explode")

plt.show()

# =====================================================
# 5. shadow # Adds a shadow behind the pie chart.
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    shadow=True,
    autopct="%1.1f%%"
)

plt.title("Pie Chart with Shadow")

plt.show()

# =====================================================
# 6. startangle # Rotates the starting position of the pie chart.

# Default starting angle = 0°
# =====================================================

plt.figure(figsize=(6,6))

plt.pie(
    survival.values,
    labels=labels,
    startangle=90,
    autopct="%1.1f%%"
)

plt.title("Pie Chart using Start Angle")

plt.show()
```

# Module 7 - Box Plot

DataSet - Employee Salary Dataset Tropics

* boxplot()
* outliers
* notch
* patch_artist

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("/content/employee_salary.csv")

# =====================================================
# 1. Basic boxplot() # boxplot() is a Matplotlib function used to create a box plot.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(df["income"])

plt.title("Basic Box Plot")

plt.show()

# =====================================================
# 2. Outliers # An Outlier is a value that is much larger or much smaller than the other data values.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    showfliers=True
)

plt.title("Box Plot with Outliers")

plt.show()

# =====================================================
# 3. notch #The notch adds a small indentation around the median.

# It helps compare the medians of different datasets.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    notch=True
)

plt.title("Notched Box Plot")

plt.show()

# =====================================================
# 4. patch_artist # patch_artist=True allows you to fill the box with color.
# =====================================================

plt.figure(figsize=(8,5))

plt.boxplot(
    df["income"],
    patch_artist=True,
    boxprops=dict(facecolor="skyblue")
)

plt.title("Colored Box Plot")

plt.show()
```

#1.Mean
average value of all values Formula Mean = ∑x/n

1.∑x = Sum of all observation

2.n = Number of observation

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# =====================================
# Mean of All Numerical Columns
# =====================================

print("\nMean of Numerical Columns")
print(df.mean(numeric_only=True))

# =====================================
# Mean of Individual Columns
# =====================================

print("\nAverage Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())
print("Average Experience:", df["Experience"].mean())
```

#2.Mode
The Mode is the value that appears most frequently

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# ============================================
# Calculate Mode
# ============================================

print("\nMode of Each Column")
print(df.mode())

# ============================================
# Mode of Individual Columns
# ============================================

print("\nMode of Age")
print(df["Age"].mode())

print("\nMode of Salary")
print(df["Salary"].mode())

print("\nMode of Experience")
print(df["Experience"].mode())

# ============================================
# First Mode Value (if multiple modes exist)
# ============================================

print("\nFirst Mode of Experience:", df["Experience"].mode()[0])

```
