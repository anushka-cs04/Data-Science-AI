#Main Data Structures in Pandas


#1.Creating Series
Aseries is a one-dimentional labeled array that stores a single column of data.

```python
import pandas as pd
marks=pd.Series([85,90,78,92,88])
print(marks)
```

Creating a series of strings

```python
import pandas as pd
names=pd.Series(["Aman","Ravi","Priya","Neha"])
print(names)
```

Creating a series with custom index

```python
import pandas as pd
marks=pd.Series(
    [85,90,78],
    index=["Maths","Science","English"]
)
print(marks)
```

Create a Series from Dictionary

```python
import pandas as pd
student = pd.Series([
    "Name"
])
```

Access Value in a Series

```python
import pandas as pd
marks = pd.Series([85,90,78,92])
print(marks[0])
print(marks[2])
```

Access values using custom Index

```python
import pandas as pd
marks = pd.Series(
    [85,90,78],
    index=["Maths","Science","English"]
)
print(marks["Science"])
```

Print Series information

```python
import pandas as pd
marks=pd.Series([85,90,78])

print(type(marks))
print(marks.dtype)
print(marks.size)
```

#Building Dataframe

```python
import pandas as pd
student={
    "Name":["Aman","Ravi","Priya","Neha"],
    "A"
}
```

Create Data frame with custom index

```python
import pandas as pd
student={
    "Name":["Aman","Ravi","Priya"],
    "Marks":[85,90,78]
}
df= pd.DataFrame(student,index=["S1","S2","S3"])
print(df)
```

create a data frame from a List of Lists

```python
import pandas as pd
data=[
    ["Aman", 20,"Delhi"],
    ["Ravi", 21, "Lucknow"],
    ["Priya", 19,"Noida"]
]

df=pd.DataFrame(data,columns=["Name","Age","City"])
print(df)
```

Create DataFrame from a List of Dictionary

```python
import pandas as pd
data=[
    {"Name": "Aman", "Age":20},
    {"Name": "Ravi", "Age":21},
    {"Name": "Priya", "Age":19},
]
df=pd.DataFrame(data)
print(df)
```
