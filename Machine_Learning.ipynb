#**Introduction to Machine Learning**
Machine Learning is the process of training a computer system using data so that it can make predictions or decision without being explicitly programed for every task.
# Real Life Examples
1. Netflix recommends movies based on your watching history.
2. YouTube suggests videos according to your interests.
3. Amazon/Flipkart recommend products you may want to buy.
4. Google Maps predicts traffic conditions.
5. Gmail automatically filters spam emails.
6. Face unlock on smartphones recognizes your face.
7. ChatGPT generates responses by learning patterns from huge amounts of text.
