```python
import pandas as pd

df = pd.read_csv("/content/student.csv")

df.info() # displays a complete summary of the data frame
print(df.shape) # returns the number of rows and columnns
print(df.columns) # shows all columns
print(df.index) # displays rows laber (index).
print(df.dtypes) # shows the data type of each column
print(df.size) # returns the total number of elements
print(df.ndim) # returns the number of dimensions.
print(df.memory_usage) # displays memory used by each column.
print(df.describe()) # provides statistical summary of numeric columns.
print(df.describe(include='all')) # shows statics for all columns (numeric nd categorical)
print(df.count()) # counts non-mising values in each column
print(df.isnull().sum()) # counts mising values
print(df.nunique()) # shows the number of unique values in each column.
print(df.value_counts()) # counts unique rows.
```

```python
import pandas as pd

df = pd.read_csv("/content/student.csv")

print(df["sex"]) # Selecting a Single Column
df[["school", "address"]] # selecting multiple columns

#Selecting Rows using Index

print(df.iloc[0]) # first row
print(df.iloc[0:5]) # multiple rows
print(df.iloc[-5:]) # last five rows

## Selecting Rows and Columns using loc[]

#loc[] uses row labels (index names) and column names.
# syntax -> df.loc[row_label, column_name]
print(df.loc[0]) # first row
print(df.loc[0:5]) # multiple rows
print(df.loc[-5:]) # last five rows

#Selecting Rows and Columns using iloc[]

#iloc[] uses integer positions.
#syntax -> df.iloc[row_index, column_index]

print(df.iloc[0]) #first row
print(df.iloc[0:5]) #multiple rows
print(df.iloc[-5:]) #last five rows

# Boolean Indexing

#Boolean Indexing returns rows where the condition is True.

# Age greater than 30
df[df["age"] > 30]

# Age less than 30
df[df["age"] < 30]
# city is banglore
df[df["address"] == "Bangalore"]

# education is master
df[df["school"] == "Masters"]

# Filtering Data

df[df["address"] == "Bangalore"] # single condition
df[(df["address"] == "Pune") & # multiple conditions(AND) (
   (df["age"] > 30)]

df[(df["address"] == "Pune") | # multiple conditions(OR)
   (df["age"] > 30)]

df[~(df["address"] == "New Delhi")] # NOT condition (changed from Department == Sales)

df[df["address"].isin(["Bangalore", "Pune"])] # using isin()

df[df["age"].between(25, 35)] # using(between)

df[df["school"].str.contains("Bachelor", case=False)] # using str.contains()
# using query
df.query("age > 30")
df.query("address == 'Pune'")
df.query("age > 30 and studytime > 3")

df.nlargest(5, "age") # using nlargest()
df.nsmallest(5, "age") # using nsmallest()
```

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Priya", np.nan],
    "Age": [22, np.nan, 25, 28, 24],
    "City": ["Delhi", "Mumbai", np.nan, "Lucknow", "Delhi"],
    "Salary": [50000, 60000, np.nan, 75000, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
print(df.isnull()) #. used to identify missing value
print(df.isnull().sum()) # count missing value
print(df.notnull()) #. used to identify missing value
# fill with a constant
filled = df.fillna("Unknown")
print(filled)
# fill numeric coulmns
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)
# fill multiple columns
df.fillna({
    "Name": "Not Available",
    "City": "Unknown",
    "Salary": 0
}, inplace=True)

print(df)

df.dropna() # removes missing values
# df.replace(old_value, new_value) # used to replace speific values (not limited to missing values)
```
